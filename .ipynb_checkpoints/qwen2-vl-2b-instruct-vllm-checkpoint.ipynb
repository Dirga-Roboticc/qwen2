```python
# !pip install git+https://github.com/huggingface/transformers vllm==0.6.1
# !pip install git+https://github.com/huggingface/transformers@21fac7abba2a37fae86106f87fcf9974fd1e3830 accelerate
# !pip install qwen-vl-utils[decord]
from transformers import AutoProcessor
from vllm import LLM, SamplingParams
from qwen_vl_utils import process_vision_info

MODEL_PATH = "Qwen/Qwen2-VL-2B-Instruct"

llm = LLM(
    model=MODEL_PATH,
    limit_mm_per_prompt={"image": 10, "video": 10},
)

sampling_params = SamplingParams(
    temperature=0.1,
    top_p=0.001,
    repetition_penalty=1.05,
    max_tokens=256,
    stop_token_ids=[],
    skip_special_tokens=False,
)
```

```python
messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {
        "role": "user",
        "content": [
            {
                "type": "image",
                "image": "./tes.png",
                # "min_pixels": 224 * 224,
                # "max_pixels": 1280 * 28 * 28,
            },
            {"type": "text", "text": "find search icon of the left sidebar of visual studio code and Generate the caption in English with grounding:"},
        ],
    },
]

# For video input, you can pass following values instead:
# "type": "video",
# "video": "<video URL>",

processor = AutoProcessor.from_pretrained(MODEL_PATH)
prompt = processor.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True,
)
image_inputs, video_inputs = process_vision_info(messages)

mm_data = {}
if image_inputs is not None:
    mm_data["image"] = image_inputs
if video_inputs is not None:
    mm_data["video"] = video_inputs

llm_inputs = {
    "prompt": prompt,
    "multi_modal_data": mm_data,
}
```

```python
# !pip install pillow matplotlib

import re
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

def extract_bbox(text):
    # Try different patterns
    patterns = [
        r'<\|box_start\|>\((.*?)\),\((.*?)\)<\|box_end\>',
        r'\(([\d,]+)\),\(([\d,]+)\)',
        r'box: \(([\d,]+)\),\(([\d,]+)\)',
        r'\((\d+),(\d+)\),\((\d+),(\d+)\)'  # New pattern for (x1,y1),(x2,y2)
    ]
    
    for pattern in patterns:
        match = re.search(pattern, text)
        if match:
            if len(match.groups()) == 2:
                top_left = tuple(map(int, match.group(1).split(',')))
                bottom_right = tuple(map(int, match.group(2).split(',')))
            elif len(match.groups()) == 4:
                x1, y1, x2, y2 = map(int, match.groups())
                top_left = (x1, y1)
                bottom_right = (x2, y2)
            return top_left, bottom_right
    
    return None

def draw_bbox(image_path, bbox):
    # Open the image
    img = Image.open(image_path)
    draw = ImageDraw.Draw(img)
    
    # Draw the bounding box
    draw.rectangle(bbox, outline="red", width=2)
    
    return img

outputs = llm.generate([llm_inputs], sampling_params=sampling_params)
generated_text = outputs[0].outputs[0].text
print("Generated text:")
print(generated_text)
print("\n" + "="*50 + "\n")

bbox = extract_bbox(generated_text)
if bbox:
    print(f"Bounding box coordinates: {bbox}")
    
    # Draw bounding box on the image
    img_with_bbox = draw_bbox("./tes.png", [bbox[0], bbox[1]])
    
    # Display the image with bounding box
    plt.figure(figsize=(10, 10))
    plt.imshow(img_with_bbox)
    plt.axis('off')
    plt.title("Image with Bounding Box")
    plt.show()
else:
    print("No bounding box found in the output.")
    print("Please check the generated text above and adjust the extraction patterns if needed.")
```
